```python
# This file was produced by the NASA Exoplanet Archive  http://exoplanetarchive.ipac.caltech.edu
# Thu Jun  3 20:55:58 2021
df = pd.read_csv("nasa_exoplanets_original.csv")
```

```python
df = df[["rowid", "pl_name", "hostname", "pl_letter", "sy_snum", "sy_pnum", "discoverymethod", "disc_year", "disc_locale",
         "disc_facility", "soltype", "pl_orbper", "pl_orbsmax", "pl_rade", "pl_orbeccen",  
         "pl_orbincl", "pl_tranmid", "pl_tsystemref", "ttv_flag", "pl_imppar", "pl_trandep", "pl_trandur", "pl_ratdor", "pl_ratror",
         "st_teff", "st_rad", "st_mass",
         "st_met", "st_metratio", "st_logg", "st_dens", "ra", "dec", "sy_pm", "sy_pmra",
         "sy_pmdec", "sy_dist", "sy_plx", "sy_bmag", "sy_vmag", "sy_jmag", "sy_hmag", "sy_kmag", "sy_w1mag", "sy_w2mag", "sy_w3mag", 
         "sy_w4mag", "sy_gaiamag", "sy_tmag", "sy_kepmag"]]
```

```python
vals = df["disc_locale"].tolist()
levels = list(set(vals))
freqs = [vals.count(i) for i in levels]
df_freq = pd.DataFrame({"level": levels,
                        "freq": freqs})
df_freq
```

```python
df.loc[(df["disc_locale"] == "Multiple Locale"), ["disc_locale"]] = "Multiple Locales"
vals = df["disc_locale"].tolist()
levels = list(set(vals))
freqs = [vals.count(i) for i in levels]
df_freq = pd.DataFrame({"level": levels,
                        "freq": freqs})
df_freq
```

```python
vals = df["st_metratio"].tolist()
levels = list(set(vals))
freqs = [vals.count(i) for i in levels]
df_freq = pd.DataFrame({"level": levels,
                        "freq": freqs})
df_freq
```

```python
df.loc[(df["st_metratio"] == "[Fe/H["), ["st_metratio"]] = "[Fe/H]"
vals = df["st_metratio"].tolist()
levels = list(set(vals))
freqs = [vals.count(i) for i in levels]
df_freq = pd.DataFrame({"level": levels,
                        "freq": freqs})
df_freq
```
